```python
# Packages

# To make data frames
import pandas as pd

# Numerical calculations
import numpy as np
```

```python
data = [['midnight', 'red', 25, '2000-01-01'], 
        ['midnight', 'red', '', '2000-01-02'],
        ['midnight', 'red', '', '2000-01-03'],
        ['midnight', 'red', '', '2000-01-04'],
        ['midnight', 'blue', 23, '2000-01-04'],
        ['midnight', 'blue', 27, '2000-01-05'],
        ['midnight', 'red', 26, '2000-01-05'],
        ['midnight', 'red', '', '2000-01-06'],
        ['midnight', 'red', '', '2000-01-07'],
        ['midnight', 'red', '', '2000-01-08'],
        ['noon', 'red', 20, '2000-01-01'], 
        ['noon', 'red', '', '2000-01-02'],
        ['noon', 'red', '', '2000-01-03'],
        ['noon', 'red', '', '2000-01-04'],
        ['noon', 'blue', 18, '2000-01-04'],
        ['noon', 'blue', 23, '2000-01-05'],
        ['noon', 'red', 21, '2000-01-05'],
        ['noon', 'red', '', '2000-01-06'],
        ['noon', 'red', '', '2000-01-07'],
        ['noon', 'red', '', '2000-01-08'],]
df = pd.DataFrame(data, columns=['time', 'color', 'value', 'date'])
df
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.sort_values(['color', 'time', 'date'], inplace=True, ignore_index=True)
df = df.replace('', np.NaN)
df
```

```python
df['value'] = df.groupby(['color', 'time'])['value'].transform(lambda x: x.ffill())
df
```

```python
df['raverage'] = df.groupby(['color', 'time']).rolling(2, min_periods=1)['value'].mean().reset_index(drop=True) 
```
